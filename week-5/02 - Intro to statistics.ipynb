### Variance and standard deviation

In addition to knowing where a distribution is *centered*, it is also important to know how *variable* the data is, that is, how *spread* it is. Two statistics, the variance and the standard deviation, are useful in this case:

$$Var = \sigma^2 = \frac{\sum (x_i - \mu) ^ 2}{N}$$

$$Std = \sigma = \sqrt{Var}$$

From [Wikipedia:](https://en.wikipedia.org/wiki/Variance) In probability theory and statistics, variance is the expected squared deviation of a random variable from its mean. Informally, it measures the distance of a set of numbers from their mean value.

(**Note**: The above expressions refer to the Variance of an entire population, where $\mu$ is the population mean. When estimating the Variance or std from a sample, the correct formula is *unbiased* is:
$$\hat{Var} = \hat\sigma^2 =\frac{\sum (x_i - \bar{x}) ^ 2}{N-1}$$

### Asymmetry (Skewness)

The relative position of the `mean`, the `median` and the `mode` of the distribution gives us an idea of the `skewness` of our data

```python
import pandas as pd
def outliers (lst):
    pd_series = pd.Series(lst)
    
    q3 = pd_series.quantile(.75)
    q1 = pd_series.quantile(.25)
    iqr = q3 - q1
    upper_bound = q3 + iqr * 1.5 
    lower_bound = q1 - iqr * 1.5
    
    outliers_list=[]
    for i in pd_series:
        if i > upper_bound:
            outliers_list.append(i)
        elif i < lower_bound:
            outliers_list.append(i)
            
    return outliers_list
```

```python
def give_me_the_outliers (series):
    
    outliers = []
    
    # 1. Range
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    
    iqr = q3-q1
    
    lower_fence = q1 - iqr*1.5
    upper_fence = q3 + iqr*1.5
    
    # 2. Values outside of that
    
    for element in series.values:
        if element > upper_fence or element < lower_fence:
            outliers.append(element)

    
    return outliers
```

### Outliers / Outliers

Boxplots typically represent outliers as data points outside the range:

$$\big[ Q_1 - 1.5 * (Q_3 - Q_1 ), Q_3 + 1.5 * (Q_3 - Q_1 ) \big]$$

This is the standard method of deciding if you are an outlier, although outliers are always going to be a problem.

The $(Q_3 - Q_1 ) := IQR$ value is known as the "interquartile range".


**Note:** There are other ways to mark observations as outliers, and the $1,5$ value can be changed to any non-negative $k$.
Depending on the distribution you have, you can calculate it asymmetrically.
If you don't say anything to the bloxplot, it calculates it for you like this.

### Covariance

[Wikipedia:](https://en.wikipedia.org/wiki/Covariance)

In probability theory and statistics, covariance is a measure of the joint variability of two random variables. If the largest values of one variable correspond mainly to the largest values of the other variable, and the same occurs with the values smaller, (that is, the variables tend to show a similar behavior), the covariance is positive In the opposite case, when the larger values of one variable correspond mainly to the smaller values of the other, (that is, variables tend to show the opposite behavior), the covariance is negative.
**The sign of the covariance shows, therefore, the trend in the linear relationship between the variables**.







The formula for the covariance between two variables ($j$, and $k$) is similar to that for the variance:

$$q_{jk}=\frac{1}{N-1}\sum_{i=1}^{N}\left(  x_{ij}-\bar{x}_j \right)  \left( x_{ik}-\bar{x}_k \right)$$

**Note:** can you see that if $j=k$ the covariance is just the variance?

### Pearson Correlation

The magnitude of the covariance is not easy to interpret because it is not normalized and therefore depends on the magnitudes of the variables. However, the normalized version of the covariance, the Pearson correlation coefficient, shows by its magnitude the strength of the linear relationship. The formula is

$$\rho_{X,Y} = \frac{Cov(X,Y)}{\sigma_X \sigma_Y}$$


where $\sigma_X$ and $\sigma_Y$ are the respective standard deviations.

Pearson's correlation coefficient is a statistic that measures the linear correlation between two variables X and Y. It has a value between +1 and -1. A value of +1 is full positive linear correlation, 0 is no linear correlation, and -1 is full negative linear correlation.


**WARNING!** The fact that the correlation (or covariance) is zero does not mean that there are no patterns

An alternative correlation coefficient known as [Spearman's correlation coefficient](https://en.wikipedia.org/wiki/Spearman%27s_rank_correlation_coefficient) solves this problem by computing Pearson's correlation for the ranks . The sign of the Spearman correlation indicates the direction of the association between X (the independent variable) and Y (the dependent variable). If Y tends to increase as X increases, Spearman's correlation coefficient is positive. If Y tends to decrease as X increases, Spearman's correlation coefficient is negative. A Spearman correlation of zero indicates that there is no tendency for Y to increase or decrease as X increases. Spearman's correlation increases in magnitude as X and Y get closer to being perfectly monotone functions of each other. When X and Y are perfectly monotonically related, the Spearman correlation coefficient becomes 1.

For more information, see [this post](https://towardsdatascience.com/clearly-explained-pearson-v-s-spearman-correlation-coefficient-ada2f473b8) in towardsdatascience.
